```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/sample_data/population.csv')
df.head()
```

```python
df.info()
```

```python
#drop
df.drop(['Country Name'], axis=1 ,inplace=True)
```

```python
# Remove duplicate rows based on all columns
df = df.drop_duplicates()

```

```python
# Check for duplicate indices
duplicate_indices = df.index.duplicated()

# If any duplicates exist, print them
print(duplicate_indices.any())
```

```python
# Check if the DataFrame has a multi-index
print(df.index)
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
Country_Code =df['Country Code'].head(16)
```

```python
Country_Code
```

```python
bar_width = 0.25
index = np.arange(len(Country_Code))

# Plotting the data
fig, ax = plt.subplots()

bar1 = ax.bar(index, df['1960'].head(16), bar_width, label='y1960', color='black')
bar2 = ax.bar(index + bar_width, df['2022'].head(16), bar_width, label='y2022', color='blue')

# # Labels and title
ax.set_xlabel('Country Code')
ax.set_ylabel('Population')
ax.set_title('Population Distribution over 1960 and 2022')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(Country_Code)
ax.legend()

# Display the plot

plt.tight_layout()
plt.show()
```

```python
values_to_delete = ['ABW', 'AND','ASM','ATG']

indexes_to_delete = df[df['Country Code'].isin(values_to_delete)].index

df.drop(indexes_to_delete, inplace=True)
```
